```python
import folium
import pandas
```

```python
data = pandas.read_csv("Volcanoes.txt")
lat = list(data["LAT"])
lon = list(data["LON"])
elev = list(data["ELEV"])
name = list(data["NAME"])

def color_producer(elevation):
    if elevation < 1000:
        return 'green'
    elif 1000 <= elevation < 3000:
        return 'orange'
    else:
        return 'red'
```

```python
map = folium.Map(location=[38.58, -99.09], zoom_start=5, tiles="Mapbox Bright")

fgv = folium.FeatureGroup(name="Volcanoes")

for lt, ln, nm, el in zip(lat, lon, name,elev):
    fgv.add_child(folium.map.Marker(location=[lt,ln],popup=folium.Popup(nm+": "+str(el))))
map.add_child(fgv)
```
