```python
def create_circle(event,x,y,flags,params):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),thickness=10)
        
img = cv2.imread(os.path.join(data_path,"dog_backpack.png"))

cv2.namedWindow("myDog")

cv2.setMouseCallback("myDog", create_circle)

while True:
    cv2.imshow("myDog",img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
